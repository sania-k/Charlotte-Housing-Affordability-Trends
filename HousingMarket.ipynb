```python
import pandas as pd
import numpy as np

df = pd.read_csv("Census_Housing_Value_Block_Groups.csv")  

print(df.duplicated().sum())
print(df.isna().sum())
# Buckets for housing values
value_cols_low = [
    'lessthan10k','a10kto14999','a15kto19999','a20kto24999',
    'a25kto29999','a30kto34999','a35kto39999','a40kto49999',
    'a50kto59999','a60kto69999','a70kto79999','a80kto89999',
    'a90kto99999','a100kto124999','a125kto149999'
]
value_cols_mid = [
    'a150kto174999','a175kto199999','a200kto249999',
    'a250kto299999','a300kto399999'
]
value_cols_high = [
    'a400kto499999','a500kto749999','a750kto999999',
    'a1Mto1499999','a1500000to1999999','a2MorMore'
]

all_value_cols = value_cols_low + value_cols_mid + value_cols_high
# Convert values to numbers and handle zeros
for col in all_value_cols + ['houses', 'Median_Housing_Value', 'MedianGrossRent']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df[['houses','Median_Housing_Value','MedianGrossRent']] = \
    df[['houses','Median_Housing_Value','MedianGrossRent']].replace(0, np.nan)

df['Affordable'] = df[value_cols_low].sum(axis=1)
df['Middle']     = df[value_cols_mid].sum(axis=1)
df['HighEnd']    = df[value_cols_high].sum(axis=1)
# Create affordability index
df['Affordability_Index'] = df['Affordable'] / df['houses']
df['Affordability_Index'] = df['Affordability_Index'].fillna(0)
df.head()
```

```python
# Affordability Category based on Affordability_Index
bins = [0, 0.33, 0.66, 1]
labels = ['Affordable', 'Middle', 'HighEnd']
df['Affordability_Index_Category'] = pd.cut(df['Affordability_Index'], bins=bins, labels=labels)

# Housing Value Brackets
housing_bins = [0, 150000, 300000, 500000, float('inf')]
housing_labels = ['<150k', '150k-300k', '300k-500k', '>500k']
df['Housing_Value_Range'] = pd.cut(df['Median_Housing_Value'], bins=housing_bins, labels=housing_labels)

# Rent Level Categories
rent_bins = [0, 800, 1500, 2500, float('inf')]
rent_labels = ['Low', 'Moderate', 'High', 'Very High']
df['Rent_Level'] = pd.cut(df['MedianGrossRent'], bins=rent_bins, labels=rent_labels)

# Percentages of unit types
df['Total_Units'] = df['Affordable'] + df['Middle'] + df['HighEnd']
df['pct_affordable'] = df['Affordable'] / df['Total_Units']
df['pct_middle'] = df['Middle'] / df['Total_Units']
df['pct_highend'] = df['HighEnd'] / df['Total_Units']

# Weighted Housing Value (for aggregation later)
df['Weighted_Housing_Value'] = df['Median_Housing_Value'] * df['houses']

# Affordable units per house
df['Affordable_per_House'] = df['Affordable'] / df['houses']

# Simplified Neighborhood / Tract Labels
# Extract block group or tract info from 'geoname'
df['Neighborhood'] = df['geoname'].str.split(',').str[0]

# Log transformations to handle skew
df['Log_Housing_Value'] = np.log1p(df['Median_Housing_Value'])
df['Log_MedianRent'] = np.log1p(df['MedianGrossRent'])

# Combined affordability score for ranking
df['Affordability_Score'] = df['pct_affordable'] - df['pct_highend']

# Save the enhanced dataset
# df.to_csv("enhanced_housing_data.csv", index=False)
```
