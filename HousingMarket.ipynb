```python
import pandas as pd
import numpy as np

df = pd.read_csv("Census_Housing_Value_Block_Groups.csv")  

print(df.duplicated().sum())
print(df.isna().sum())
# Buckets for housing values
value_cols_low = [
    'lessthan10k','a10kto14999','a15kto19999','a20kto24999',
    'a25kto29999','a30kto34999','a35kto39999','a40kto49999',
    'a50kto59999','a60kto69999','a70kto79999','a80kto89999',
    'a90kto99999','a100kto124999','a125kto149999'
]
value_cols_mid = [
    'a150kto174999','a175kto199999','a200kto249999',
    'a250kto299999','a300kto399999'
]
value_cols_high = [
    'a400kto499999','a500kto749999','a750kto999999',
    'a1Mto1499999','a1500000to1999999','a2MorMore'
]

all_value_cols = value_cols_low + value_cols_mid + value_cols_high
# Convert values to numbers and handle zeros
for col in all_value_cols + ['houses', 'Median_Housing_Value', 'MedianGrossRent']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df[['houses','Median_Housing_Value','MedianGrossRent']] = \
    df[['houses','Median_Housing_Value','MedianGrossRent']].replace(0, np.nan)

df['Affordable'] = df[value_cols_low].sum(axis=1)
df['Middle']     = df[value_cols_mid].sum(axis=1)
df['HighEnd']    = df[value_cols_high].sum(axis=1)
# Create affordability index
df['Affordability_Index'] = df['Affordable'] / df['houses']
df['Affordability_Index'] = df['Affordability_Index'].fillna(0)
df.head()
```
